3.2 Merging the telcom dataset with the GeoCell dataset

```
#Merging the final telecom dataset with the geocell dataset
#final_dataset = pd.merge(df_2, geocell, on='SITE_ID', how='outer')
#final_dataset  = pd.merge(df_2, geocell,left_on='SITE_ID', right _on='SITE_CODE')

final_dataset = pd.merge( df_2, geocell, left_on='SITE_ID' , right_on='SITE_CODE',validate='m:1', how='inner')
final_dataset

```

```
#splitting the date and time 
df_date = final_dataset['DATETIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(' '))]))
df_date.columns = ['TIME', 'DATE']
df_date


```

```
df_time = df_date['TIME'].apply(lambda x: pd.Series([i for i in reversed(x.lower().split(':'))]))
df_time.columns = ['s','m','h']
df_time['h'] = pd.to_numeric(df_time['h'])
df_time['m'].unique()
```

```
#merging to the final dataset
final_dataset = df_date.merge(final_dataset, left_index=True, right_index=True)
final_dataset

#merging the splitted time to the final dataset
final_dataset = df_time.merge(final_dataset, left_index=True, right_index=True)
final_dataset
```

```
#droping DATETIME column
final_dataset = final_dataset.drop('DATETIME', axis=1)
final_dataset
```
